```python
# Imports
import matplotlib.pyplot as plt
import seaborn as sns

from uniharmony import make_multisite_classification, verbosity
from uniharmony.plot import plot_2d_projection


sns.set_theme(style="whitegrid")
verbosity("warning")
```

### Let's simulate data with only Effects of Site (EoS) or only real signal.

```python
X, y, sites = make_multisite_classification(n_classes=5, n_sites=5, signal_strength=5, site_effect_strength=0)
plot_2d_projection(X, sites)


X, y, sites = make_multisite_classification(n_classes=5, n_sites=5, signal_strength=0, site_effect_strength=5)
fig, axes = plot_2d_projection(X, sites)
plt.show()
```

```python
X, y, sites = make_multisite_classification(n_classes=5, n_sites=5, signal_strength=10, site_effect_strength=0)
plot_2d_projection(X, sites, y, method="tsne")

X, y, sites = make_multisite_classification(n_classes=5, n_sites=5, signal_strength=0, site_effect_strength=10, random_state=23)
plot_2d_projection(X, sites, y, method="tsne")
plt.show()
```

```python
X, y, sites = make_multisite_classification(n_classes=5, n_sites=5, n_features=1200, signal_strength=10, site_effect_strength=0)
plot_2d_projection(X, sites, y, method="pca")


X, y, sites = make_multisite_classification(n_classes=5, n_sites=5, n_features=120, signal_strength=0, site_effect_strength=10)
fig, axes = plot_2d_projection(X, sites, y, method="pca")
plt.show()
```
