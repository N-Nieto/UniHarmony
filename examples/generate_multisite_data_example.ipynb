# Generate data

```python
# Imports
import numpy as np

from uniharmony import make_multisite_classification
```

```python
# Quick test of the function
print("Testing simulate_multi_site_data...")

# Test 1: Binary classification
print("\nTest 1: Binary classification (default)")
X1, y1, sites1 = make_multisite_classification(
    n_sites=3,
    n_samples=300,
    n_features=20,
    verbose=True,
    site_effect_homogeneous=False,
)
print(f"Shapes: X={X1.shape}, y={y1.shape}, sites={sites1.shape}")
print(f"Unique classes: {np.unique(y1)}")
print(f"Unique sites: {np.unique(sites1)}")

# Test 2: Multi-class classification
print("\nTest 2: Multi-class classification")
X2, y2, sites2 = make_multisite_classification(
    n_sites=2,
    n_samples=200,
    n_features=15,
    n_classes=3,
    verbose=True,
    site_effect_homogeneous=True,
)
print(f"Shapes: X={X2.shape}, y={y2.shape}, sites={sites2.shape}")
print(f"Unique classes: {np.unique(y2)}")

# Test 3: Custom class balance
print("\nTest 3: Custom class balance")
balance = [0.8, 0.2]  # 80% class 1 for both sites
X3, y3, sites3 = make_multisite_classification(
    n_sites=2, n_samples=100, balance_per_site=balance, verbose=True
)
print(f"Class distribution: {np.bincount(y3)}")
```
